
### Recursive Algorithm for Factorial Calculation

The task is to calculate the factorial of a number `n (denoted as n! )`, which is the product of all positive integers up to `n` .

#### Definition:
-  `n! = n * (n - 1)!`
- **Base Case**:
  - If `n = 0`  or `n = 1` , return `1` because `0! = 1! = 1` .

---

### Recursive Algorithm (Pseudocode)

```text
function factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

---

### Step-by-Step Example for `factorial(4)`

#### Function Calls:
```
factorial(4) → factorial(3) → factorial(2) → factorial(1) → factorial(0)
```

#### Step-by-Step Evaluation:
1. **Base Case**:
   - `factorial(0) = 1` (returns 1)
2. **Returning Values**:
   - `factorial(1) = 1` (returns 1)
   - `factorial(2) = 2 * 1 = 2`
   - `factorial(3) = 3 * 2 = 6`
   - `factorial(4) = 4 * 6 = 24`

Result:  
4! = 24

---

### C++ Code for Factorial Calculation (Recursive)

```cpp
#include <iostream>
using namespace std;

// Recursive function to calculate factorial
int factorial(int n) {
    // Base case: if n is 0 or 1
    if (n == 0 || n == 1)
        return 1;
    
    // Recursive case: n! = n * (n-1)!
    return n * factorial(n - 1);
}

int main() {
    int n;
    cout << "Enter a number (n): ";
    cin >> n;

    // Call the recursive function
    int result = factorial(n);
    cout << n << "! = " << result << endl;

    return 0;
}
```

---

### Example for `factorial(5)`

#### Function Calls:
```
factorial(5) → factorial(4) → factorial(3) → factorial(2) → factorial(1) → factorial(0)
```

#### Step-by-Step Evaluation:
1. **Base Case**:
   - `factorial(0) = 1` (returns 1)
2. **Returning Values**:
   - `factorial(1) = 1`
   - `factorial(2) = 2 * 1 = 2`
   - `factorial(3) = 3 * 2 = 6`
   - `factorial(4) = 4 * 6 = 24`
   - `factorial(5) = 5 * 24 = 120`

Result:  
5! = 120
