### Recursive Algorithm for Fibonacci Series

The Fibonacci series is a sequence of numbers in which each number is the sum of the two preceding ones, usually starting with 0 and 1.

#### Definition:
- The Fibonacci sequence is defined as follows:
  - F(0) = 0 and F(1) = 1
  - For n ≥ 2: F(n) = F(n-1) + F(n-2)
    
---

### Recursive Algorithm (Pseudocode)

```text
function fibonacci(n):
    if n == 0:
        return 0
    else if n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

---

### Step-by-Step Example for `fibonacci(5)`

#### Function Calls:
```
fibonacci(5) → fibonacci(4) → fibonacci(3) → fibonacci(2) → fibonacci(1) → fibonacci(0)
```

#### Step-by-Step Evaluation:
1. `fibonacci(0) = 0` (Base case)
2. `fibonacci(1) = 1` (Base case)
3. `fibonacci(2) = fibonacci(1) + fibonacci(0) = 1 + 0 = 1`
4. `fibonacci(3) = fibonacci(2) + fibonacci(1) = 1 + 1 = 2`
5. `fibonacci(4) = fibonacci(3) + fibonacci(2) = 2 + 1 = 3`
6. `fibonacci(5) = fibonacci(4) + fibonacci(3) = 3 + 2 = 5`

Result:  
The 5th Fibonacci number is F(5) = 5.

---

### C++ Code for Fibonacci Calculation (Recursive)

```cpp
#include <iostream>
using namespace std;

// Recursive function to calculate Fibonacci number
int fibonacci(int n) {
    // Base cases
    if (n == 0 && n== 1)
        return 0;
    
    // Recursive case: F(n) = F(n-1) + F(n-2)
    return fibonacci(n - 1) + fibonacci(n - 2);
}

int main() {
    int n;
    cout << "Enter the position (n): ";
    cin >> n;

    // Call the recursive function
    int result = fibonacci(n);
    cout << "Fibonacci number at position " << n << " is " << result << endl;

    return 0;
}
```

### Example for `fibonacci(6)`

#### Function Calls:
```
fibonacci(6) → fibonacci(5) → fibonacci(4) → fibonacci(3) → fibonacci(2) → fibonacci(1) → fibonacci(0)
```

#### Step-by-Step Evaluation:
1. `fibonacci(0) = 0`
2. `fibonacci(1) = 1`
3. `fibonacci(2) = 1`
4. `fibonacci(3) = 2`
5. `fibonacci(4) = 3`
6. `fibonacci(5) = 5`
7. `fibonacci(6) = 5 + 3 = 8`

Result:  
The 6th Fibonacci number is F(6) = 8.
