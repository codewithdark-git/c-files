### Recursive Algorithm for Power Calculation

The task is to calculate `x^n` (i.e., `x` raised to the power of `n`) using recursion.

#### Base Case:
- If n == 0, return `1` because any number raised to the power of 0 is 1.

#### Recursive Case:
- If n > 0 return `x * power(x, n-1)` (recursive call with reduced exponent).


---

### Recursive Algorithm (Pseudocode)

```text
function power(x, n):
    if n == 0:
        return 1 
    else:
        return x * power(x, n - 1)
```

---

### Step-by-Step Example for `power(2, 3)`

#### Function Calls:
```
power(2, 3) → power(2, 2) → power(2, 1) → power(2, 0)
```

#### Step-by-Step Evaluation:
- `power(2, 0) = 1` (Base case)
- `power(2, 1) = 2 * 1 = 2`
- `power(2, 2) = 2 * 2 = 4`
- `power(2, 3) = 2 * 4 = 8`

Result:  
2^3 = 8

### Optimized Recursive Algorithm for Power (C++ Code)

```cpp
#include <iostream>
using namespace std;

// Recursive function to calculate power
int power(int x, int n) {
    // Base case: if exponent is 0
    if (n == 0)
        return 1;

    else {
        return x * power(x, n - 1);
    }
}

int main() {
    int x, n;
    cout << "Enter base (x): ";
    cin >> x;
    cout << "Enter exponent (n): ";
    cin >> n;

    // Call the recursive function
    int result = power(x, n);
    cout << x << "^" << n << " = " << result << endl;

    return 0;
}
```
